```python
import matplotlib.pyplot as plt
```

```python
X = [1,2,3] # 원하는 그래프를 그릴때 X랑 Y를 따로 관리하는 것으로 접근하면 됨.
Y = [1,3,4]
plt.plot(X, Y, "o--") 
plt.show()
```

```python
# 수식을 넣어 그래프를 그리고 싶을때(회귀 직선이나 곡선, 다항회귀 그림을 그릴때) 
X = range(-20, 20) # 원하는 그래프를 그릴때 X랑 Y를 따로 관리하는 것으로 접근하면 됨.
Y = [x**2+1 for x in X]
plt.plot(X, Y, "o--") 
plt.show()
```

```python
# 점을 직접 넣는 방법
X = range(0, 5) # 원하는 그래프를 그릴때 X랑 Y를 따로 관리하는 것으로 접근하면 됨.
Y = [2*x+1 for x in X]
plt.plot(X, Y)
DataX =[1,3,4]            #DataX,Y는 원본 데이터 
DataY =[2.9, 7.1, 9.3]
plt.plot(DataX, DataY, "o") #원본 데이터를 넣을때는 "o" 마커를 넣어 표시하는게 일반적이다.
plt.show()
```

```python
# 점을 직접 넣는 방법(정수 범위를 벗어난)
import math as m        # y = sin(2*pi*t) 주기적인 값을 도출할때
X = range(0, 40) 
Y = [m.sin(x) for x in X]
plt.plot(X, Y, "o--")   # 주기를 6으로 갖는 그래프가 나옴. ## 단 조금 거칠어 보인다는 흠이 있음 왜 그럴까?
plt.show()                                                 ## 점의 갯수가 모자라 보임.
```

```python
import numpy as np
X = np.linspace(0,1,4)  # 0에서부터 1까지 4개를 만들어보자라는 의미 range 보다 조금 더 디테일함.
print(X)
```

```python
# linspace(range 함수의 정수만 넣을 수 있는 부분을 보완함.)
X = np.linspace(0, 2*np.pi, 100) # 100등분 해보자
Y = np.sin(X)
plt.plot(X,Y);plt.show()
```

```python
# 히스토그램 (랜덤은 중간값을 많이 내주는 걸 볼 수 있음)
data = np.random.randn(1000)
plt.hist(data, bins = 15)  # bin 큰 상자(막대의 갯수를 정해줌, 표준정규분포란 데이터가 없는게 아닌 연속적으로 고르게 분포함을 뜻함)
plt.show()                 # 따라서 표준정규분포를 나타낼 수 있는 적절한 bin값을 넣어 줘야함.
```

```python
dataA=np.random.standard_normal((100,2))
dataB=np.random.standard_normal((100,2))
dataC=np.random.standard_normal((100,2))
dataA += np.array((-3,-3))
dataB += np.array((4,4))
dataC += np.array((-1,4))
plt.scatter(dataA[:,0], dataA[:,1])
plt.scatter(dataB[:,0], dataB[:,1]) # color = 1 흰색 0.5 회색 / 색상표를 검색해 "#~~~" 형태로 넣어주면 해당 색깔이 반영됨.
plt.scatter(dataC[:,0], dataC[:,1])
```

```python
dataA=np.random.standard_normal((100,2))
dataB=np.random.standard_normal((100,2))
dataC=np.random.standard_normal((100,2))
dataA += np.array((-3,-3))
dataB += np.array((2,4))
dataC += np.array((-1,4))
plt.scatter(dataA[:,0], dataA[:,1], edgecolor ="0")
plt.scatter(dataB[:,0], dataB[:,1], edgecolor ="0") # edgecolor의 장점으로 겹치는 데이터를 잘 나타낼 수 있음.
plt.scatter(dataC[:,0], dataC[:,1], edgecolor ="0")
```

```python
# 컬러 다양화 라이브러리
import matplotlib.cm as cm
N = 256
angle = np.linspace(0, 8*2*np.pi, N)
radius = np.linspace(5, 1, N)
X = radius*np.cos(angle)
Y = radius*np.sin(angle)
plt.scatter(X, Y, c=angle, cmap=cm.gray) # cm.Gray # cm.hsv(무지개) # cm.Spectral
plt.show()
```

```python
X = np.linspace(-10, 10, 100) # -10 ~ 10 까지 100개
Y = [2*x**2+1 for x in X]
Y2 = [3*x**2+5 for x in X]
Y3 =  [0.5*x**2+3 for x in X]
plt.plot(X,Y, linestyle = "solid")
plt.plot(X,Y2, linestyle = "dashed")
plt.plot(X,Y3, linestyle = "dashdot")
plt.show()
```

```python
# 누적 막대 그래프(총계를 중심으로 비교하기가 용이함)
N=4
A = np.random.random(N) # 4개의 랜덤 수
B = np.random.random(N)
X = np.arange(N)
plt.bar(X, A, color = "r")
plt.bar(X, A+B, bottom = A)
```

```python
N=4
A = np.array([1,3,4,2]) # 4개의 랜덤 수
B = np.array([3,2,1,2])
C = np.array([4,3,2,2])
X = np.arange(N)
plt.bar(X, A, color = "r")
plt.bar(X, A+B, color = "b", bottom = A)
plt.bar(X, A+B+C, bottom = A+B) # C 의 나머지 구간은 A + B 라는 점 꼭 주의하기
```

```python
N=4
A = np.array([1,3,4,2]) # 4개의 랜덤 수
B = np.array([3,2,1,2])
C = np.array([4,3,2,2])
X = np.arange(N)
plt.bar(X, A, color = "r",hatch = "/")
plt.bar(X, A+B, color = "b", hatch= "x",bottom = A)
plt.bar(X, A+B+C, bottom = A+B, hatch= "|") # C 의 나머지 구간은 A + B 라는 점 꼭 주의하기
```

```python
# 마커 심화
A = np.random.standard_normal((100,2))
B = np.random.standard_normal((100,2))

A += np.array((-1,-2))
B += np.array((0.5,1))
plt.scatter(A[:,0], A[:,1], color ="k", marker= "o") # ^, o, x, v,<,>,s(네모),p(오각형),P(플러스),*,X,d(다이아몬드)
plt.scatter(B[:,0], B[:,1], color ="k", marker= "x")
```
